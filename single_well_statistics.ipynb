```python
import pandas as pd
from sklearn import datasets
import seaborn as sns
import matplotlib.pyplot as plt
from glob import glob
import warnings
import matplotlib.pyplot as plt
warnings.filterwarnings("ignore")
well_list = glob(r'USROP*.csv', recursive=False)
# df = pd.read_csv('USROP_A 0 N-NA_F-9_Ad.csv')
df = pd.read_csv(well_list[0])
df.drop('Unnamed: 0.1',axis=1, inplace=True)
df.drop('Mud Density In g/cm3',axis=1, inplace=True)
df.drop('Diameter mm',axis=1, inplace=True)
df.drop('Unnamed: 0',axis=1, inplace=True)
df.dropna(axis=0, how='any', inplace=True)
df
```

```python
# correlation matrix
sns.pairplot(df)
```

```python
plt.figure(figsize=(12,12))
corr = df.corr()
sns.heatmap(corr, xticklabels=corr.columns, yticklabels=corr.columns,cmap="YlGnBu",annot=True)
```

```python
plt.figure(figsize=(15,15))
for i in range(len(df.columns)):
    plt.subplot(3,4,i+1)
    plt.title(df.columns.values[i])
    sns.distplot(df[df.columns.values[i]])
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.plot(df['Measured Depth m'], df['Rate of Penetration m/h'])
font = {'family' : 'Arial',
'weight' : '400',
'size'   : 16,
}
plt.ylabel('Rate of Penetration m/h',font)
plt.xlabel('Measured Depth m',font)
```
